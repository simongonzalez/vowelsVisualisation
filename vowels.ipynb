```python
import plotly
import ipywidgets as widgets
import plotly.plotly as py
import plotly.figure_factory as ff
from plotly.graph_objs import *
import plotly.graph_objs as go
import functools
import pandas as pd
import numpy as np
from plotly.widgets import GraphWidget
from ipywidgets import interact
import seaborn as sns

plotly.offline.init_notebook_mode()
from plotly.offline import iplot
```

```python
#https://stackoverflow.com/questions/43288550/iopub-data-rate-exceeded-when-viewing-image-in-jupyter-notebook
df = pd.read_csv("fulldata_p.csv")
df2 = df.copy()
df2['ethnicity'] = 'Anglo'
df2['class'] = 'LWC'
df2['stress'] = 1
toChange = ["name","sex","ethnicity","ses","vowel","word"]

for col in toChange:
    df2[col] = df2[col].astype('category')
    df2[col + '_plot'] = df2[col].cat.codes

df2.head(5)
```

```python
sampleN = widgets.IntSlider(
    value=70,
    min=1,
    max=110,
    step=1,
    description='Sample',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

column_headers_dropdown = widgets.Dropdown(
    options={column: column for column in df2.columns},
    value='sex_plot',
    description='Grouping',
)

def view_image(sampleN, column_headers_dropdown):
    data_in = df2.sample(n=sampleN)

    iplot({
        'data': [go.Scatter(
            x=data_in.F2,
            y=data_in.F1,
            mode='markers',
            marker=dict(
                size='16',
                color = data_in[column_headers_dropdown],
                colorscale='Viridis',
                showscale=True)
        )],
        "layout": go.Layout(
            xaxis=dict(
                autorange='reversed'
            ),
            yaxis=dict(
                autorange='reversed'
            ),
            title="Sample Bar Chart")
    })
    
interact(view_image, sampleN = sampleN, column_headers_dropdown = column_headers_dropdown)

```
